```python
import pandas as pd
import plotly.express as px
```

```python
file = pd.read_csv("datasets\energyco2.csv")
df_energyco2 = pd.DataFrame(file)
```

```python
df_energyco2 = df_energyco2.drop(columns="Unnamed: 0")
```

```python
df_energyco2.head()
```

```python
df_energyco2.info()
```

```python
df_energyco2
```

```python
data = df_energyco2[['Country', 'Energy_type', 'Year', 'CO2_emission']]

d = (
    data.loc[data['Energy_type'] == 'all_energy_types', ['Country', 'Year', 'CO2_emission']
    ].groupby(['Year', 'Country', 'CO2_emission']).mean()
)
```

```python
d = d.reset_index()
```

```python
d = d.query("Country != 'World'")
```

```python
d = d.sort_values(by='CO2_emission',ascending=False)
```

```python
fig = px.line(d,x='Year', y='CO2_emission',color='Country')
fig.show()
```

```python
df_co2 = df_energyco2.query("Energy_type == 'all_energy_types' & Country != 'World' & Country != 'Former U.S.S.R.' & Country !='Germany, West'").groupby(by='Country').agg({"CO2_emission":"mean"}).sort_values(by='CO2_emission', ascending=False)
```

```python
df_c = df_co2.head(10)
```

```python
fig = px.pie(df_c,values='CO2_emission',names=df_c.index)
fig.update_layout(title='Contribucion por pais en emisiones CO2, top 10')
fig.show()
```
